```python
N = 100
```

```python
xs = [i for i in range(N)]
main_batch =  [i for i in range(N)]
```

```python
def E(y, a, b, batch = xs): # функция сумм квадратов расстояний 
    return sum([(y[i] - a*i - b) ** 2 for i in batch])
```

```python
def Eda(y, a, b, batch = xs):
    return -2 * sum([(y[i] - a*i - b) * i for i in batch])
```

```python
def Edb(y, a, b, batch = xs):
    return -2 * sum([(y[i] - a*i - b) * 1 for i in batch])
```

```python
sigma = 3   # стандартное отклонение наблюдаемых значений
at = 0.5     # теоретическое значение параметра k
bt = 2       # теоретическое значение параметра b

f = np.array([at*i+bt for i in xs])
y = np.array(f + np.random.normal(0, sigma, N))
```

```python
plt.scatter(xs, y, s=1)
plt.plot(xs, f, c = 'black')
```

```python
a_plt = np.arange(-1, 2, 0.1)
b_plt = np.arange(0, 15, 0.1)
E_plt = np.array([[E(y, a, b) for a in a_plt] for b in b_plt])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection='3d')

a, b = np.meshgrid(a_plt, b_plt)
ax.plot_surface(a, b, E_plt, color='y', alpha=0.5)
ax.scatter(0,0, E(y,0,0), c='red')  
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('E')   
```

```python
lmd1 = 0.000001
lmd2 = 0.0005

lamds=np.array([lmd1, lmd2]) 
    
```

```python
Eda(y, 0, 0, xs)
```

```python
Edb(y, 0, 0,  xs)
```

```python
def grad (y, params, batch):
    return np.array([Eda(y, params[0], params[1], batch),  Edb(y, params[0], params[1], batch)])
```

```python
def gradient_descent(aa, bb, batch_size = N):  # по умолчанию полный батч, т.е. просто градиентный спуск
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    k = 0
    val_pred = E(y, aa, bb)
    val_new = val_pred + 0.05 
    while(abs(val_pred - val_new)> 0.03 and k < 100):
        k = k + 1
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            params = params - lamds * grad(y, params, main_batch[batch * batch_size : (batch+1) * batch_size])
        arr.append(params)
        val_pred = val_new
        val_new = E(y, params[0], params[1])
    print(k)
    return arr
```

```python
def app(dots): 
    ass = [i[0] for i in dots]
    bs = [i[1] for i in dots]
    
    a_plt = np.arange(0, max(ass), 0.1)
    b_plt = np.arange(0, max(bs), 0.1)
    E_plt = np.array([[E(y, a, b) for a in a_plt] for b in b_plt])
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection='3d')

    for i in dots:
        ax.scatter(i[0], i[1], E(y, i[0], i[1]), c='red')  
    
    a, b = np.meshgrid(a_plt, b_plt)
    ax.plot_surface(a, b, E_plt, color='y', alpha=0.5)

    ax.set_xlabel('a')
    ax.set_ylabel('b')
    ax.set_zlabel('E')   
    plt.show()

    fnew = np.array([dots[-1][0]*i+dots[-1][1] for i in xs])
    print(dots[-1][0],dots[-1][1])

    plt.scatter(xs, y, s=1)
    plt.plot(xs, f, c = 'black')
    plt.plot(xs, fnew, c = 'red')
    return dots
```

```python
def norm(y):
    '''
        mi ...x....ma
        0 ....x.....1
    '''
    ma = max(y) #1
    mi = min(y) #0
    ras = ma - mi 
    return [(x - mi) / ras for x in y]
```

```python
GAMMA = 0.5
```

```python
def grad_momentum (y, params, batch, pred):
    return GAMMA*pred + lamds * grad(y, params, batch) 

def momentum_gradient_descent(aa, bb, batch_size = N):  
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    pred = 0
    for i in range(50):
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            pred = grad_momentum(y, params, main_batch[batch * batch_size : (batch+1) * batch_size], pred)
            params = params - pred
        arr.append(params)
    return arr
```

```python
d_mom = momentum_gradient_descent(0,0)
memory_usage()
```

```python
app(d_mom)
```

появляется инерционное движение за счет истории о пред градиентах 

Нестеров. градиент считается не в самой точке, а в следующей точке

```python


def grad_nesterov (y, params, batch, pred):
    now = GAMMA*pred
    params = params - now
    return now + lamds * grad(y, params, batch) 

def nesterov_gradient_descent(aa, bb, batch_size = N):  
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    pred = 0
    for i in range(50):
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            pred = grad_nesterov(y, params, main_batch[batch * batch_size : (batch+1) * batch_size], pred)
            params = params - pred
        arr.append(params)
    return arr
```

```python
d_nesterov = nesterov_gradient_descent(0,0)
memory_usage()
```

```python
app(d_nesterov)
```

Adagard некоторые перематры могут быстрее достигать своего оптимума. шаг зависит от величины флуктуации, чем они больше, тем меньше шаг 

```python
lamds = np.array([0.05, 0.05])
```

```python
eps = 0.03

def gradient_descent_Adagard (aa, bb, batch_size = N):  
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    pred = 0
    for i in range(50):
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            gr = grad(y, params, main_batch[batch * batch_size : (batch+1) * batch_size])
            pred = pred + gr*gr
            params = params - lamds * gr / np.sqrt(pred+eps)
            
        arr.append(params)
    return arr
```

```python
d_Adagard = gradient_descent_Adagard(0,0)
memory_usage()
```

```python
app(d_Adagard)
```

RMSProp

```python
def gradient_descent_RMSProp (aa, bb, batch_size = N):  
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    pred = 0
    for i in range(50):
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            gr = grad(y, params, main_batch[batch * batch_size : (batch+1) * batch_size]) 
            pred = GAMMA * pred + (1-GAMMA) *gr*gr
            params = params - lamds * gr / np.sqrt(pred+eps)
            
        arr.append(params)
    return arr
```

```python
d_RMSProp = gradient_descent_RMSProp(0,0)
memory_usage()
```

```python
app(d_RMSProp)
```

Adam - использует сглаженные версии среднего и среднеквадратического градиентов

```python
beta1 = 0.95
beta2 = 0.99
```

```python
 def gradient_descent_Adam (aa, bb, batch_size = N):  
    arr = [np.array([aa,bb])]
    params = np.array([aa,bb])
    num_steps = N // batch_size
    main_batch = [ i for i in range(N)]
    m = 0
    v = 0
    for i in range(50):
        np.random.shuffle(main_batch)
        for batch in range(num_steps):
            
            gr = grad(y, params, main_batch[batch * batch_size : (batch+1) * batch_size])  
            m = beta1*m + (1-beta1)*gr
            v = beta2*v + (1-beta2)*gr*gr
            params = params - lamds * m /np.sqrt(v+eps)
            
        arr.append(params)
    return arr
```

```python
d_Adam = gradient_descent_Adam(0,0)
memory_usage()
```

```python
app(d_Adam)
```
